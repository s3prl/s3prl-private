```python
import torch
import torchaudio
from torch.nn.utils.rnn import pad_sequence

from s3prl import Output
from s3prl.nn import NNModule

class MyUpstream(NNModule):
    def __init__(self, num_ceps: int = 13):
        super().__init__()
        self.num_ceps = num_ceps

    @property
    def downsample_rate(self):
        return 160  # This is the default downsample_rate of MFCC

    @property
    def input_size(self):
        return 1

    @property
    def output_size(self):
        return self.num_ceps

    @property
    def num_hidden_state(self):
        return 1

    def forward(self, x, x_len):
        mfccs = []
        for _x, _x_len in zip(x, x_len):
            mfcc = torchaudio.compliance.kaldi.mfcc(_x[:_x_len].view(1, -1), num_ceps=self.num_ceps)
            mfccs.append(mfcc)
        return Output(
            output=[torch.nn.utils.rnn.pad_sequence(mfccs, batch_first=True)],
            output_len=(torch.Tensor(x_len) / self.downsample_rate).round().long(),
        )
    
model = MyUpstream()
x_len = [16000, 32000, 48000]
x = pad_sequence([torch.randn(l, 1) for l in x_len], batch_first=True)
output, output_len = model(x, x_len).slice(2)
print(output[0].shape)
print(output_len)
```

## Is the comparison valid?

We know that both HuBERT and WavLM outperform our original upstream MFCC.
Now, does WavLM really outperforms HuBERT according to the above experinment?
According to the [doc](http://140.112.21.12:8000/index.html), we know that

| name | model | pre-training data |
| - | - | - |
| hubert | 12-layer transformer blocks | LibriSpeech 960 hours |
| wavlm | 12-layer transformer blocks | LibriLight 60k hours + Gigaspeech 10k hours + VoxPopuli 24k hours = 94k hours |

Many times, there are lots of differences across without side-by-side ablation, however, at least there are some most important factors affecting a SSL pre-trained model's performance you must check:
- **model size (parameter scale)**: more important than the specific model architecture (like Transformer or Conformer)
- **batch size (computation scale)**: this can cause **HUGE** differences according to the table below from HuBERT's paper



- **unlabeled speech amount (data scale)**: we can further explore this below

Most of the time, you can get huge performance boost by simply scale the algorithm up without changing the design.
Hence, when checking the paper, it is really important to make sure two rows are comparible (at least on these well-known most imporatant factors), to conclude that one algorithm wins the another.

### Speaker Identification

Next, we examine SSL's capability on capturing speaker information, and how they are robust against few-shot learning, just like it has been shown in wav2vec 2.0 that 10 mins of transcription is sufficient to train an ASR system.

```python
from typing import List
from s3prl import Workspace, newdict
from s3prl.problem.superb.sid import SuperbSID

def low_resource_sid(
    root_workspace: Workspace,
    upstream_name: str,
    upstream_cfg: dict,
    ratios: List[int],
    total_steps: int = 50,
):
    for ratio in ratios:
        subspace = root_workspace / upstream_name / f"ratio_{ratio}"
        SuperbSID.run_stages(
            workspace=subspace,
            setup=dict(
              corpus=newdict(
                  _cls="mini_voxceleb1",
                  dataset_root="../mini_voxceleb1"
              ),
              train_datapipe={
                  "-1": dict(
                      _cls="BalancedRatioSampler",
                      target_name="label",
                      ratio=ratio,
                  )
              },
              upstream=newdict(
                  **upstream_cfg,
              ),
            ),
            train=dict(
              optimizer=dict(
                  lr=1.0e-2,
              ),
              trainer=dict(
                  total_steps=total_steps,
                  log_step=10,
                  save_step=50,
                  eval_step=50,
                  eval_batch=50,  # For faster turnaround, just use the first 50 batches for validation
              )
            ),
        )
```

```python
low_resource_sid(workspace, "HuBERT", dict(_cls="S3PRLUpstream", name="hubert"), ratios=[1.0, 0.5, 0.2], total_steps=100)
```

```python
low_resource_sid(workspace, "Fbank", dict(_cls="S3PRLUpstream", name="fbank_no_cmvn"), ratios=[1.0, 0.5, 0.2], total_steps=500)
```

```python
def result(upstream_name: str, ratio: float):
    return (workspace / upstream_name / f"ratio_{ratio}")["test_metrics"]["accuracy"]

import matplotlib.pyplot as plt
plt.plot([result('Fbank', 1.0), result('Fbank', 0.5), result('Fbank', 0.2)], label="Fbank")
plt.plot([result('HuBERT', 1.0), result('HuBERT', 0.5), result('HuBERT', 0.2)], label="HuBERT")
plt.legend()
```

## Spoken Language Understanding (SLU)

Next, we can also test the SSL models ability on SLU task, where the model need to classify the intent from an utterance, so that a SLU system can automate the service for the users. Like in an fightlight booking system, the virtual assistant needs to classify your spoken departure location/time, and arrival location/time. Or, for a smart home system like Alexa, it needs to identify from your command (like, turn on the light in the kitchen) that the action is "turn on", the object is "light", and the location is "kitchen", so that it can automate this action for you. Furthermore, noise-robusness is critical for this task as you won't want to say the command again and again in the real-world senario. (Like, people in the house are talking, the fan is spinning, and your cat is meowing... etc) Hence, we also add noise for this task here.

We use the **Fluent Speech Command** dataset for the intent classification task, and augment with the small **Nonspeech** noise dataset.

```python
import shutil
from s3prl import Workspace, newdict
from s3prl.problem.superb.ic import SuperbIC

def noisy_mini_fsc(root: Workspace, upstream_name: str, upstream_cfg: dict, noisy: bool = True):
    # The number of the dictionary key is the priority in the datapipe
    # The datapipe with the smaller number well be executed first
    if noisy:
        effects = {
            "90": dict(
                _cls=AdditiveNoise,
                noise_paths_hook=dict(
                    _cls=find_files,
                    directory="/home/leo/d/datasets/noise_data/Nonspeech"
                ),
                snrs=(3, 9),
            ),
            "100": dict(
                _cls=Reverberation,
                wav_name="wav_noisy",
            )
        }
        subspace = root / "ic" / upstream_name / "noisy"
    else:
        effects = dict()
        subspace = root / "ic" / upstream_name / "clean"
        
    if subspace.is_dir():
        shutil.rmtree(str(subspace))

    SuperbIC.run_stages(
        workspace=subspace,
        setup=dict(
            corpus=newdict(
                _cls="mini_fsc",
                dataset_root="../mini_fsc",
            ),
            upstream=newdict(
                **upstream_cfg,
            ),
            train_datapipe=effects,
            valid_datapipe=effects,
            test_datapipe=effects,
        ),
        train=dict(
          optimizer=dict(
              lr=1.0e-2,
          ),
          trainer=dict(
              total_steps=500,
              log_step=10,
              save_step=500,
              eval_step=500,
          ),
        ),
    )
    print(f"{upstream_name}: {subspace['test_metrics']}")
```

```python
noisy_mini_fsc(workspace, "fbank", dict(_cls="S3PRLUpstream", name="fbank_no_cmvn"), noisy=False)
```

```python
noisy_mini_fsc(workspace, "fbank", dict(_cls="S3PRLUpstream", name="fbank_no_cmvn"), noisy=True)
```

```python
noisy_mini_fsc(workspace, "hubert", dict(_cls="S3PRLUpstream", name="hubert"), noisy=False)
```

```python
noisy_mini_fsc(workspace, "hubert", dict(_cls="S3PRLUpstream", name="hubert"), noisy=True)
```

We can find that, even under the noisy SSL features still work, to a certain level. At least they are still much better than our traditional choise. However, we can also find that, currently they are not robust enough against noises.

## Simple challenge: Which upstream can be robust enough against few-shot and noises?

```python
from librosa.util import find_files

from s3prl import Workspace, newdict
from s3prl.problem.superb.ic import SuperbIC
from s3prl.dataset.fewshot import RatioSampler
from s3prl.dataset.effects import AdditiveNoise, Reverberation

def challenge_fsc(
    upstream_cfg: dict,
    total_steps: int = 500,
    log_step: int = 10,
    save_step: int = 500,
    eval_step: int = 500,
    lr: float = 1.0e-2,
) -> Workspace:
    subspace = Workspace()
    
    effects = {
        "-1": dict(
            _cls=RatioSampler,
            ratio=0.5,
        ),
        "90": dict(
            _cls=AdditiveNoise,
            noise_paths_hook=dict(
                _cls=find_files,
                directory="/home/leo/d/datasets/noise_data/Nonspeech"
            ),
            snrs=(3, 9),
        ),
        "100": dict(
            _cls=Reverberation,
            wav_name="wav_noisy",
        ),
    }

    SuperbIC.run_stages(
        workspace=subspace,
        setup=dict(
            corpus=newdict(
                _cls="mini_fsc",
                dataset_root="../mini_fsc",
            ),
            upstream=newdict(
                **upstream_cfg,
            ),
            train_datapipe=effects,
            valid_datapipe=effects,
            test_datapipe=effects,
        ),
        train=dict(
          optimizer=dict(
              lr=lr,
          ),
          trainer=dict(
              total_steps=total_steps,
              log_step=log_step,
              save_step=save_step,
              eval_step=eval_step,
              eval_batch=5,
          ),
        ),
    )
    return subspace
```

```python
from s3prl.nn.upstream import S3PRLUpstream

subspace = challenge_fsc(
    dict(
        _cls=S3PRLUpstream,
        name="wav2vec2",
    ),
    lr=1e-2,
)
print(subspace["test_metrics"])
```
